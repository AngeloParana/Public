```python
import pandas as pd
df = pd.json_normalize(response.json(),"value")
display(df)
```

```python
#Get all datasets

response = requests.get(url = "https://api.powerbi.com/v1.0/myorg/datasets",headers=header)

dataset_arr = response.json()['value']

all_refreshes = []

for id in dataset_arr :
    myID = id["id"]
    datasetname = id['name']
    refresh_response = requests.get(url="https://api.powerbi.com/v1.0/myorg/datasets/" + myID  +"/refreshes",headers = header)
    df_refreshes = pd.json_normalize(refresh_response.json(),'value')
    df_refreshes["datasetid"] = myID
    df_refreshes["datasetname"]=datasetname

    all_refreshes.append(df_refreshes)

final_df = pd.concat(all_refreshes, ignore_index=True)
display(final_df)


```
